```python
from pathlib import Path
import os
import requests
import wget
import json
import pandas as pd
import ast
import io
from ast import literal_eval
URL = "https://sparse.tamy.edu/"
DIR = os.path.join(str(Path.home()),"src", "tools",
        "suitesparse-matrix-collection-website", "db", "collection_data",
        "matrices")
files = {}
filepaths = {}
for (dirpath, dirnames, filenames) in os.walk(DIR):
    dir_name = (os.path.basename(dirpath))
    files[dir_name] = []
    filepaths[dir_name] = []
    for filename in filenames:
        if (filename.endswith('.rb')):
            files[dir_name].append(filename)
            filepaths[dir_name].append(os.path.join(dirpath, filename))


            
            
def parse_file(input_file):
    dict_input = {}
    with open (input_file, "r") as fin:
        Line = fin.readline()
        while Line:
            if Line.strip() not in ["{", "}", "\n"]:
                Line = Line.strip()[:-1]
                Line = Line.replace("'", "")
                if(Line.endswith(",")):
                    Line = Line[:-1]
                elements = Line.strip().split(':')
                elements = [i.replace('"', '') for i in elements]
                elements = [i.strip() for i in elements]
                if(len(elements)>1):
                    dict_input[elements[0]] = elements[1]
                #
            Line = fin.readline()
    df_input = pd.DataFrame.from_dict(dict_input, orient = 'index')
    folder_name = input_file.split("/")[-2]
    filename = input_file.split("/")[-1][:-3]
    df_input['filename'] = filename
    df_input['folder_name'] = folder_name
    return df_input

    #print(dict_input)
#         Lines = fin.readlines()
#         Lines = [i.strip() for i in Lines if i not in ['\n']]
#         s = "".join(Lines)
#         s = s.replace("\'", "\"")
#         s = s.replace(',",', '"')
#         d = ast.literal_eval(json.dumps(s))

all_data = pd.DataFrame()
for k, paths in filepaths.items():
    for path in paths:
        #print(path)
        df_file = parse_file(path)
        all_data = pd.concat([all_data, df_file])
        break
    #break

#print (filepaths)
all_data
```

```python
DATA_DIR = os.path.join(os.getcwd(), "data")
png_files = []
for (dirpath, dirnames, filenames) in os.walk(DATA_DIR):
    dir_name = (os.path.basename(dirpath))
    for filename in filenames:
        if (filename.endswith('.png')):
            png_files.append(os.path.join(dirpath, filename))
```

```python
%matplotlib inline
import numpy as np
#from IPython.display import Image 
import matplotlib.pyplot as plt
from PIL import Image 

def pil_fn(f):
    im = Image.open(f).convert('L')
    im = im.resize((32,32))
    im_np = np.array(im).ravel()
    return im_np

png_np_array = [pil_fn(f) for f in png_files]
# Currently using PIL to conver to grayscale
# Alternative : 
# import numpy as np
# import matplotlib.pyplot as plt
# import matplotlib.image as mpimg

# def rgb2gray(rgb):
#     return np.dot(rgb[...,:3], [0.2989, 0.5870, 0.1140])

# img = mpimg.imread('image.png')     
# gray = rgb2gray(img)    
# plt.imshow(gray, cmap=plt.get_cmap('gray'), vmin=0, vmax=1)
# plt.show()

#plt.imshow(im)
#display(im)
```

```python
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
print (np.shape(png_np_array))
tsne = TSNE().fit_transform(png_np_array)
tx, ty = tsne[:,0], tsne[:,1]
tx = (tx-np.min(tx)) / (np.max(tx) - np.min(tx))
ty = (ty-np.min(ty)) / (np.max(ty) - np.min(ty))
```

```python
# THis code automatically embeds the images into a bigger image to visualize how the embedding looks like
def pil_fn1(f):
    im = Image.open(f)
    im = im.resize((32,32))
    return im

png_array = [pil_fn1(f) for f in png_files]
width = 4000
height = 3000
max_dim = 100
full_image = Image.new('RGB', (width, height))
for idx, f in enumerate(png_files):
    tile = Image.open(f)
    rs = max(1, tile.width / max_dim, tile.height / max_dim)
    tile = tile.resize((int(tile.width / rs),
                        int(tile.height / rs)),
                       Image.ANTIALIAS)
    full_image.paste(tile, (int((width-max_dim) * tx[idx]),
                            int((height-max_dim) * ty[idx])))
```
